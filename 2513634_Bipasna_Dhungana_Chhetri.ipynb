#6. Adding HDI Category Column:

• Create a new column HDI Category that classifies each country into one of the four official
Human Development Index groups. The classification should be based on the HDI value for the
year 2022. Use the following categories and thresholds defined by the United Nations Development
Programme (UNDP):

HDI Category HDI Range (hdi)
Low < 0.550
Medium 0.550 – 0.699
High 0.700 – 0.799
Very ≥ 0.800

After creating this new column:
• verify that all countries are classified correctly,
• ensure the updated dataframe includes the new category column.
• Save the final dataframe as HDI category added.csv and include this file in your final
submission.

```python
# Function to classify HDI categories based on UNDP thresholds:

def classify_hdi(value):
    if value < 0.550:
        return "Low"
    elif value >= 0.550 and value <= 0.699:
        return "Medium"
    elif value >= 0.700 and value <= 0.799:
        return "High"
    else:
        return "Very High"

# Creating new column 'HDI Category'
hdi_2022_df["HDI Category"] = hdi_2022_df["hdi"].apply(classify_hdi)

# Checking if all countries are classified
hdi_2022_df["HDI Category"].value_counts()

# Saving the final dataframe
hdi_2022_df.to_csv("/content/drive/MyDrive/2513634_Bipasna_Dhungana_Chhetri_Final_Assignment_AI/HDI_category_added.csv", index=False)

# Displaying the first few rows to confirm new column
hdi_2022_df.head()
```

#3. Outlier Detection:

• Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the South Asia dataset
df = pd.read_csv("/content/drive/MyDrive/2513634_Bipasna_Dhungana_Chhetri_Final_Assignment_AI/HDI_SouthAsia.csv")

# Using the HDI Outliers
Q1_hdi = df["hdi"].quantile(0.25)
Q3_hdi = df["hdi"].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi

# uSING the GNI Outliers to quantile for math operation
Q1_gni = df["gross_inc_percap"].quantile(0.25)
Q3_gni = df["gross_inc_percap"].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni

# Identifying the outliers
df["Outlier"] = (
    (df["hdi"] < lower_hdi) | (df["hdi"] > upper_hdi) |
    (df["gross_inc_percap"] < lower_gni) | (df["gross_inc_percap"] > upper_gni)
)

# Viewing the outlier countries
df[df["Outlier"] == True]


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the datasets
south_asia_df = pd.read_csv("/content/drive/MyDrive/2513634_Bipasna_Dhungana_Chhetri_Final_Assignment_AI/HDI_SouthAsia_2020_2022.csv")

middle_east_df = pd.read_csv("/content/drive/MyDrive/2513634_Bipasna_Dhungana_Chhetri_Final_Assignment_AI/HDI_MiddleEast_2020_2022.csv")

# Getting the average HDI per country (2020–2022)
south_avg = south_asia_df.groupby("country")["hdi"].mean().reset_index()
middle_avg = middle_east_df.groupby("country")["hdi"].mean().reset_index()

# Top and bottom 3 countries according to question
south_top = south_avg.sort_values("hdi", ascending=False).head(3)
south_bottom = south_avg.sort_values("hdi").head(3)

middle_top = middle_avg.sort_values("hdi", ascending=False).head(3)
middle_bottom = middle_avg.sort_values("hdi").head(3)

# Combining the data for plotting graphs
combined = pd.concat([
    south_top.assign(Region="South Asia - Top"),
    south_bottom.assign(Region="South Asia - Bottom"),
    middle_top.assign(Region="Middle East - Top"),
    middle_bottom.assign(Region="Middle East - Bottom")
])

# Displaying the Bar chart
plt.figure(figsize=(10, 5))
plt.bar(combined["country"], combined["hdi"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Country")
plt.ylabel("Average HDI (2020–2022)")
plt.title("Top and Bottom HDI Performers: South Asia vs Middle East")
plt.tight_layout()
plt.show()
```

#Metric Comparisons:

• Compare the following metrics across regions using grouped bar charts:

– Gender Development Index {"gender development"}
– Life Expectancy Index {"life expectancy"}
– GNI per Capita Index {"gross inc percap"}

• Identify which metric shows the greatest disparity between regions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Loading the datasets
south_asia_df = pd.read_csv("/content/drive/MyDrive/2513634_Bipasna_Dhungana_Chhetri_Final_Assignment_AI/HDI_SouthAsia_2020_2022.csv")

middle_east_df = pd.read_csv("/content/drive/MyDrive/2513634_Bipasna_Dhungana_Chhetri_Final_Assignment_AI/HDI_MiddleEast_2020_2022.csv")

# Metrics to compare with three of them:
metrics = ["gender_development", "life_expectancy", "gross_inc_percap"]

# Calculating the mean values
south_means = south_asia_df[metrics].mean()
middle_means = middle_east_df[metrics].mean()

# X positions
x = np.arange(len(metrics))
width = 0.35

# Plot grouped bar chart to average and compare the south vs middle
plt.figure(figsize=(8, 5))
plt.bar(x - width/2, south_means, width, label="South Asia")
plt.bar(x + width/2, middle_means, width, label="Middle East")

plt.xlabel("Metrics")
plt.ylabel("Average Value (2020–2022)")
plt.title("Comparison of Development Metrics: South Asia vs Middle East")
plt.xticks(x, metrics, rotation=20)
plt.legend()

plt.tight_layout()
plt.show()
```

#HDI Disparity:

• Compute the range (max – min) of HDI for each region.

• Compute the coefficient of variation (CV = std/mean) for HDI.

• Identify which region exhibits more variation in HDI.

```python
# HDI Range to find max and min for each region
south_range = south_asia_df["hdi"].max() - south_asia_df["hdi"].min()
middle_range = middle_east_df["hdi"].max() - middle_east_df["hdi"].min()

# Mean and Standard Deviation
south_mean = south_asia_df["hdi"].mean()
south_std = south_asia_df["hdi"].std()

middle_mean = middle_east_df["hdi"].mean()
middle_std = middle_east_df["hdi"].std()

# Coefficient of Variation
south_cv = south_std / south_mean
middle_cv = middle_std / middle_mean

# Displaying the results
print("South Asia HDI Range:", south_range)
print("South Asia CV:", south_cv)

print("Middle East HDI Range:", middle_range)
print("Middle East CV:", middle_cv)
```

#Outlier Detection:

• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.

• Create scatter plots highlighting outliers in a different color.

• Discuss the significance of these outliers.

```python
#First of all lets do with south aisan outliers ditection:

import pandas as pd
import matplotlib.pyplot as plt

# Loading the datasets
south_df = pd.read_csv("/content/drive/MyDrive/2513634_Bipasna_Dhungana_Chhetri_Final_Assignment_AI/HDI_SouthAsia_2020_2022.csv")

#   Using the IQR for HDI
Q1_hdi = south_df["hdi"].quantile(0.25)
Q3_hdi = south_df["hdi"].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

hdi_lower = Q1_hdi - 1.5 * IQR_hdi
hdi_upper = Q3_hdi + 1.5 * IQR_hdi

# Using the IQR for GNI
Q1_gni = south_df["gross_inc_percap"].quantile(0.25)
Q3_gni = south_df["gross_inc_percap"].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

gni_lower = Q1_gni - 1.5 * IQR_gni
gni_upper = Q3_gni + 1.5 * IQR_gni

# Identifying the outliers
outliers = south_df[
    (south_df["hdi"] < hdi_lower) | (south_df["hdi"] > hdi_upper) |
    (south_df["gross_inc_percap"] < gni_lower) | (south_df["gross_inc_percap"] > gni_upper)
]

normal = south_df.drop(outliers.index)

# Displaying and Using the Scatter plot
plt.figure()
plt.scatter(normal["gross_inc_percap"], normal["hdi"])
plt.scatter(outliers["gross_inc_percap"], outliers["hdi"])
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("South Asia: Outliers in HDI vs GNI per Capita")
plt.show()
```
